```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/train.csv')
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df.columns
```

**We have to build a model to predict the sales price based on the square footage and the number of bathrooms and bedrooms. So we have to seperate those columns first**

```python
df=df[['LotArea','LotFrontage','TotalBsmtSF','GrLivArea','BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','SalePrice']]
df
```

```python
sns.pairplot(df)
```

```python
df.isna().sum()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X=df.drop(columns=['SalePrice'])
y=df['SalePrice']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=54)
```

```python
X_train
```

```python
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R^2 Score: {r2}')
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted')
plt.show()
```

**Hence we can say that the model is accurate to an extent**
